### Gráfico da abertura de lojas ao longo do tempo

É possível perceber uma curva mais acentuada em relação à proporção de Life.

```python
fig, ax = plt.subplots(figsize=(20, 8))

vivara = df[['Data de Abertura', 'Nº de Lojas Acumulado por Tipo', 'Tipo']].loc[df['Tipo'] == 'LOJA']
quiosqui = df[['Data de Abertura','Nº de Lojas Acumulado por Tipo', 'Tipo']].loc[df['Tipo'] == 'QUIOSQUE']
life = df[['Data de Abertura', 'Nº de Lojas Acumulado por Tipo', 'Tipo']].loc[df['Tipo'] == 'LOJA LIFE']

ax.grid(False)
ax.plot(vivara['Data de Abertura'], vivara['Nº de Lojas Acumulado por Tipo'], label = 'Loja Vivara', c = 'purple')
ax.plot(quiosqui['Data de Abertura'], quiosqui['Nº de Lojas Acumulado por Tipo'], c = 'pink', label = 'Quiosque')
ax.plot(life['Data de Abertura'], life['Nº de Lojas Acumulado por Tipo'], c = 'blue', label = 'Loja Life')

ax.set_title('Abertura de lojas ao longo do tempo', fontsize = 20)
leg = plt.legend(loc='upper left', prop={'size': 20})

plt.tight_layout(pad = 5.0);
```

### Gráfico do crescimento de life e vivara percentualmente

Podemos perceber que o número de lojas life teve um crescimento considerável após 2020, mas vinha sido mantido estável nos anos anteriores. De fato, a empresa ficou 4 anos sem abrir novas lojas do ramo, abriu apenas mais 4 lojas LIFE em 2019, e 47 lojas nos anos posteriores.

```python
fig, ax = plt.subplots(figsize=(20, 8))

vivara = df[['Data de Abertura', 'Nº de Lojas Acumulado por Tipo', 'Tipo']].loc[df['Tipo'] == 'LOJA']
life = df[['Data de Abertura', 'Nº de Lojas Acumulado por Tipo', 'Tipo']].loc[df['Tipo'] == 'LOJA LIFE']
vivara = vivara.loc[vivara['Data de Abertura'] >= '2015-08-07']

ax.grid(False)
ax.plot(vivara['Data de Abertura'], vivara['Nº de Lojas Acumulado por Tipo']/vivara['Nº de Lojas Acumulado por Tipo'].max(), label = 'Loja Vivara', c = 'purple')
ax.plot(life['Data de Abertura'], life['Nº de Lojas Acumulado por Tipo']/life['Nº de Lojas Acumulado por Tipo'].max(), c = 'blue', label = 'Loja Life')

ax.set_title('Crescimento de Life e Vivara Percentualmente', fontsize = 20)
leg = plt.legend(loc='upper left', prop={'size': 20})

plt.tight_layout(pad = 5.0);
```

### Abertura de lojas de 2015 a 2019

De 2015 a 2019, apenas 3 lojas life foram abertas e posteriormente, mais 3 lojas. Contudo, nos anos seguintes a 2020 foi aberto 47 lojas LIFE. A abertura de lojas Vivara foi constante, sendo aberta 40 lojas durante o mesmo período (2015 a 2019), e apenas 42 lojas nos anos seguintes.
Em 2020 a vivara teve a o recorde de vendas digitais, com +191,2%, que representa R$ 309,4 milhões, em vendas digitais no ano, juntamente com +41% de novos clientes.

### Público-Alvo

Vivara: lançada em 1962, é associada a uma marca tradicional, elegante e moderna destinada a clientes acima dos 30 anos de idade, que prezam por qualidade e sofisticação.

Life by Vivara: lançada em 2011, é associada a um público-alvo mais jovem, acima de 15 anos de idade, que valoriza a moda e procura peças mais descontraídas, colecionáveis e com um preço mais acessível. Inicialmente a marca contava com pulseiras e pingentes em seu portfólio, e veio adicionando coleções e outras categorias de produtos com o passar do tempo.
